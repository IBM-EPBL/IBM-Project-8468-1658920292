```python
# filling na values with mean
cdf = df.fillna(df.mean())
```

```python
cdf.isnull().sum()
```

```python
cdf.drop(columns=['WindGustDir', 'WindSpeed3pm','WindDir3pm'],axis=1,inplace=True)
```

```python
cdf['RainToday'].fillna(cdf['RainToday' ]. mode(), inplace=True)
cdf['RainTomorrow' ].fillna(cdf['RainTomorrow'].mode(), inplace=True)
```

```python
cdf.isnull().any()
```

```python
cdf = cdf.dropna(axis=0)
cdf.isnull().sum()
```

```python
sns.countplot(x = 'RainTomorrow' , data = cdf)
```

```python
cdf["RainToday"] = cdf.RainToday.apply(lambda x: 0 if x == "Yes" else 1)
cdf["RainToday"] = cdf.RainToday.apply(lambda x: 0 if x == "No" else 1)
```

```python
cdf["RainTomorrow"] = cdf.RainToday.apply(lambda x: 0 if x == "Yes" else 1)
cdf["RainTomorrow"] = cdf.RainToday.apply(lambda x: 0 if x == "No" else 1)
```

```python
cdf[["Location","RainToday"]].groupby("Location").sum().sort_values(by='RainToday',ascending=False).plot(kind='barh',stacked=True,figsize=(15,10))
plt.xlabel("Rainfall in MM",size=12)
plt.ylabel("Location",size=15)
plt.title("Annual Rainfall v/s Locations")
plt.grid(axis="x",linestyle="--")
plt.show()
```

```python
plt.figure(figsize=(15,8))
cdf.groupby("Date").sum()['RainToday'].plot(kind="line",color="r",marker=".")
plt.xlabel("Date",size=15)
plt.ylabel("RAINFALL IN MM",size=15)
plt.grid(axis="both",linestyle="-.")
plt.title("Rainfall over Years")
plt.show()
```
